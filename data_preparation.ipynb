```python
import cv2 
import os 

data_path = os.path.join('data','UBFC-RPPG-Dataset')
subjects = os.listdir(data_path)  
```

```python
import numpy as np
import cv2
print("numpy:", np.__version__)
print("cv2:", cv2.__version__)

img = np.zeros((200,200,3), dtype=np.uint8)
print("OK")
```

```python
import cv2
import urllib.request
import os 




url = "https://github.com/opencv/opencv_zoo/raw/main/models/face_detection_yunet/face_detection_yunet_2023mar.onnx"
model_path = "yunet.onnx"
if not os.path.exists(model_path):
    urllib.request.urlretrieve(url, model_path)



```

```python
import matplotlib.pyplot as plt

for subject in subjects: 
    #if there is already a roi_colors.txt file, skip the subject
    if os.path.exists(os.path.join(data_path, subject, 'roi_colors.txt')):
        print(f"roi_colors.txt already exists for {subject}, skipping...")
        continue
    print(f"Processing {subject}...")
    video_path = os.path.join(data_path, subject , 'vid.avi')
    cap = cv2.VideoCapture(video_path)

    success, img = cap.read()
    while success:

        h, w = img.shape[:2]

        detector = cv2.FaceDetectorYN.create(
            model_path,
            "",
            (w, h),
            0.6,
            0.3,
            5000
        )

        _, faces = detector.detect(img)
        face = faces[0]
        x1, y1, w, h = face[:4].astype(int) 
        
        # Forehead ROI (upper part of face box)
        forehead_x1 = x1 + int(0.2 * w)
        forehead_x2 = x1 + int(0.8 * w)
        forehead_y1 = y1
        forehead_y2 = y1 + int(0.25 * h)

           
        # Left cheek ROI
        left_cheek_x1 = x1
        left_cheek_x2 = x1 + int(0.35 * w)
        left_cheek_y1 = y1 + int(0.35 * h)
        left_cheek_y2 = y1 + int(0.75 * h)

           
        # Right cheek ROI
        right_cheek_x1 = x1 + int(0.65 * w)
        right_cheek_x2 = x1 + w
        right_cheek_y1 = y1 + int(0.35 * h)
        right_cheek_y2 = y1 + int(0.75 * h)

        #average color in each roi 
        forehead_roi = img[forehead_y1:forehead_y2, forehead_x1:forehead_x2]
        left_cheek_roi = img[left_cheek_y1:left_cheek_y2, left_cheek_x1:left_cheek_x2]
        right_cheek_roi = img[right_cheek_y1:right_cheek_y2, right_cheek_x1:right_cheek_x2]

        forehead_avg_color = np.mean(forehead_roi, axis=(0,1))
        left_cheek_avg_color = np.mean(left_cheek_roi, axis=(0,1))
        right_cheek_avg_color = np.mean(right_cheek_roi, axis=(0,1))

        merged_colors = np.concatenate([forehead_avg_color, left_cheek_avg_color, right_cheek_avg_color])
        save_dir = os.path.join(data_path,subject)
        with open(os.path.join(save_dir, 'roi_colors.txt'), 'a') as f:
            f.write(','.join(map(str, merged_colors)) + '\n')

        success, img = cap.read()
      
     
    

            
        
       
```
